```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_json("../assets/clean_reviews.json", orient='records')
data
```

```python
labels = np.array(list(map(str, data['record_label'])))
for a in np.flatnonzero(np.core.defchararray.find(labels,'|')!=-1):
    string = labels[a]
    split = np.array(string.split("|"))
    labels = np.concatenate((labels, split))
    
label = np.take(labels, np.flatnonzero(np.core.defchararray.find(labels,'|')==-1))

unique = np.unique(label)
print(len(unique))
amount = []
for name in unique:
    amount.append(np.count_nonzero(label == name))

    
amount = np.array(amount)

relevantLabels = unique[amount > 100]
relevantAmount = amount[amount > 100]
print(len(relevantLabels))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(relevantLabels, relevantAmount)
plt.show()
print(relevantLabels)
```
